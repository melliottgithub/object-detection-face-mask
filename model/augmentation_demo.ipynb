```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from PIL import Image
from lib import image_augments as aug
import math
```

## Define Augmentations

```python
def image_augmentations(img):
    augments = aug.rgb_shift(img, 25)
    augments.extend(aug.channel_shuffle(img))
    augments.extend(aug.blur(img))
    augments.extend(aug.enhancements(img, factors=(0.5, 0.5, 0.1, 0.3)))
    augments.extend(aug.rotations(img, [90,270,15]))
    augments.extend(aug.transpose(img))
    return augments
```

## Original Image

```python
image_path = 'images/train/PMM/10.jpg'
img = Image.open(image_path)
img.size
plt.imshow(img)
plt.show()
```

## Augmentations

```python
augments = image_augmentations(img)
rows = math.ceil(len(augments) / 3)
fig, ax = plt.subplots(rows, 3, figsize=(12, 12))

for i, aug_img in enumerate(augments):
    axs = ax.flat[i]
    axs.imshow(aug_img[1])
    axs.set_title(aug_img[0])
    axs.axis('off')
```
